```python
import pickle
import matplotlib.pyplot as plt
```

```python
pickle_file = 'data/GENKI_4_2020-01-31-21.59.15.pickle'
with open(pickle_file, "rb") as input_file:
    data = pickle.load(input_file)
```

```python
randidx = data['randidx']
pixels = data['pixels']
images = data['images']
labels = data['labels']
masks = data['masks']
```

```python
fig, axs = plt.subplots(1, 3, figsize=(18, 6))
test_idx = 0

print(labels[test_idx])
axs[0].imshow(images[test_idx].reshape(256, 256), cmap='bone')
axs[1].imshow(masks[test_idx].reshape(256, 256), cmap='bone')
axs[2].imshow(pixels[test_idx].reshape(4, 4), cmap='seismic')

plt.show()
```
